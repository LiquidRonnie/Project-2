```python
clean_df[['date', 'temp1', 'temp2', 'temp3', 'temp4','temp5','temp6','temp7','temp8','temp9','temp10','temp11' ]] = clean_df['rawData'].str.split('\t', 11, expand=True)
clean_df.head()
```

```python
clean_df[['year', 'mo']] = clean_df['date'].str.split('/', 2, expand=True)
clean_df.head()
```

```python
del clean_df['rawData']
clean_df.head()
```

```python
clean_df[[ 'temp1', 'temp2', 'temp3', 'temp4','temp5','temp6','temp7','temp8','temp9','temp10','temp11' ]] = clean_df[[ 'temp1', 'temp2', 'temp3', 'temp4','temp5','temp6','temp7','temp8','temp9','temp10','temp11' ]].astype(float)
```

```python
# Drop Rows with Missing Information & Count to comapir
clean_df = clean_df.dropna(how='any')
clean_df.count()
clean_df.describe()
clean_df['temp1'].dtype
clean_df.head()
```

```python
col = clean_df.loc[: , "temp1":"temp11"]
clean_df['meanTemp'] = col.mean(axis=1)
clean_df.head()
```

```python
clean_df = clean_df.drop(['temp1', 'temp2', 'temp3', 'temp4','temp5','temp6','temp7','temp8','temp9','temp10','temp11' ], axis=1)
clean_df.head()
```
